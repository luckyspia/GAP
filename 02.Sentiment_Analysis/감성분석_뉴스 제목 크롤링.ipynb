```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from tqdm import tqdm
import re
import warnings
warnings.filterwarnings(action='ignore')
```

```python
def get_news_info(keyword, Year, sDate, eDate):
    info=[]
    for page in tqdm(range(1,1001,10)):       
        url = "https://search.naver.com/search.naver?sm=tab_hty.top&where=news&query=" + keyword
        url = url + "&ds=" + str(Year) +sDate + "&de=" + str(Year) + eDate
        url = url + "&mynews=0&office_section_code=0&office_type=0&pd=3&photo=0&sort=0"
        url = url + "&start=" + str(page)
        
        html = requests.get(url)
        soup = bs(html.text, "html.parser")
        
        section=soup.find("section", class_="sc_new sp_nnews _prs_nws")
        areas=section.find_all("div", class_="news_area")

        for area in areas:
            date = area.find("span", class_="info").get_text()
            title = area.find("a", class_="news_tit").get_text()
            title=re.sub('<[^<]+?>',"",title).replace("&quot;","")
            info.append([title,date])
        
    df=pd.DataFrame(info, columns=["title","date"])
    return df        
```

```python
def get_write(df,keyword, Year):
    ft = open('c:/pydata/' + keyword + '_뉴스제목_' + str(Year) + '.txt', "a", encoding="utf-8")
    
    for row in df['title']:
        ft.write(row+"\n")
    
    ft.close()
```

```python
keyword="노인일자리"
sDate="01.01"
eDate="12.31"
for Year in range(2006, 2023):
    if Year==2022:
        eDate="05.10"
    df=get_news_info(keyword,Year,sDate, eDate)
    get_write(df,keyword,Year)
```

```python
# 전체로 통합해서 txt 파일 만들기

keyword="노인일자리"
sDate=".01.01"
eDate=".12.31"
df=pd.DataFrame()
for Year in range(2006, 2023):
    if Year==2022:
        eDate="05.10"
    news_df=get_news_info(keyword,Year,sDate, eDate)
    df=pd.concat([df, news_df])
    
ft = open('c:/pydata/' + keyword + '_뉴스제목_ALL.txt', "a", encoding="utf-8")
for row in df['title']:
    ft.write(row+"\n")
ft.close()
```
